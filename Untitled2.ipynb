```python
# Integers are whole numbers, positive or negative, without decimals.

# Assigning integer values to variables
integer_number = 10
negative_integer = -5
large_integer = 1000000

# Performing basic arithmetic operations with integers
addition_result = integer_number + negative_integer
subtraction_result = integer_number - negative_integer
multiplication_result = integer_number * 2
division_result = integer_number // 3 # Integer division (discards remainder)
modulo_result = integer_number % 3 # Modulo (returns the remainder)

# Printing the values and results
print("Integer number:", integer_number)
print("Negative integer:", negative_integer)
print("Large integer:", large_integer)
print("Addition result:", addition_result)
print("Subtraction result:", subtraction_result)
print("Multiplication result:", multiplication_result)
print("Integer division result:", division_result)
print("Modulo result:", modulo_result)

# Checking the data type of a variable
print("Data type of integer_number:", type(integer_number))
```
